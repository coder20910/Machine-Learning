# Matplotlib Practice

# Data Visualisation

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.array([1,2,3,4,5])
y = x**2 + 3
```

```
print(x,y)
```

# Data visualisation using Line

```
plt.plot(x,y)      # Plotting without styling
```

```
plt.style.use('seaborn')            # Here seaborn styling is used
plt.plot(x,y,label="Stock Price")   # Label is added to the left corner of graph
plt.title('My Line Project')        # Title is displayed above  the graph
plt.xlabel('TimeStamp')             # Name  of the 'x' axis is printed
plt.ylabel('Price')                 # Name of the 'y' axis is printed
plt.legend()                        # Legend is responsible for inner styling
plt.show()                          # It will display all the figure in noniteractive mode
```

# Data visualisation using scatter

```
plt.figure(figsize=(10,10))              # It is used to adjust the size of the graph
plt.style.use('dark_background')        # Here dark background theme is used
plt.scatter(x,y,label="Stock Price")    # Scatter will draw the graph in connecting point form
plt.title('My Scatter Project')         # Title is displayed above  the graph
plt.xlabel('TimeStamp')                 # Name  of the 'x' axis is printed
plt.ylabel('Price')                     # Name of the 'y' axis is printed
plt.legend()                            # Legend is responsible for inner styling
plt.show()          
```

# Visualisation using Bar Graph

```
plt.bar([0,2,1],[10,20,15])    # It will draw the basoc bar graph   
plt.show()
```

```
x_cordinates = np.array([0,1,2])*2
plt.bar(x_cordinates,[10,20,15],width = 0.5,label='Current Year',tick_label=["Gold", "Platinun","Silver"])
plt.bar(x_cordinates+0.5,[20,10,12],width=0.5,label='Next Year')
plt.title("Metal Price comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```
